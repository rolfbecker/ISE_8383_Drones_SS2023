# My First FFT Tests


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Load the WAV file
filename = 'Data/CantinaBand3.wav'
sample_rate, data = wavfile.read(filename)

# Take the Fourier transform
fft_data = np.fft.fft(data)

# Get the frequency values corresponding to the FFT data
fft_freq = np.fft.fftfreq(len(data), 1/sample_rate)

# Plot the wave and FFT data
fig, axs = plt.subplots(3, 1, figsize=(12,8))
axs[0].plot(data)
axs[0].set_title('Waveform')
axs[1].plot(fft_freq, np.abs(fft_data))
axs[1].set_title('FT abs')
axs[1].set_xlim([0, 1000])
#axs[1].set_xlim([0, sample_rate/2])
axs[2].plot(fft_freq, np.angle(fft_data))
axs[2].set_title('FT angle')
#axs[2].set_xlim([0, sample_rate/2])
axs[2].set_xlim([0, 1000])

plt.show()
```

# TWO SIGNALS!

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Load the WAV file
filename = 'Data/CantinaBand3.wav'
sample_rate, data = wavfile.read(filename)

data1 = data[10000:50000]
data2 = data[20000:60000]

# Take the Fourier transform
fft1 = np.fft.fft(data1)
fft2 = np.fft.fft(data2)

# Get the frequency values corresponding to the FFT data
fft_freq = np.fft.fftfreq(len(data1), 1/sample_rate)

# Plot the wave and FFT data
fig, axs = plt.subplots(4, 1, figsize=(16,8))
axs[0].plot(data1)
axs[0].plot(data2)
axs[0].set_title('Waveform')
axs[1].plot(fft_freq, np.abs(fft1))
axs[1].plot(fft_freq, np.abs(fft2))
axs[1].set_title('FT abs')
axs[1].set_xlim([0, 1000])
#axs[1].set_xlim([0, sample_rate/2])
axs[2].plot(fft_freq, np.angle(fft1))
axs[2].plot(fft_freq, np.angle(fft2))
axs[2].set_title('FT angle')
#axs[2].set_xlim([0, sample_rate/2])
axs[2].set_xlim([0, 1000])

axs[3].plot(fft_freq, np.angle(fft1/fft2))
axs[3].set_title('FT Quotient angle')
#axs[2].set_xlim([0, sample_rate/2])
axs[3].set_xlim([0, 1000])

plt.show()
```
